#### - http://quotes.toscrape.com

```python
import pandas as pd
import numpy as np
import csv
import matplotlib.pyplot as plt
```

```python
import requests
from bs4 import BeautifulSoup

url = "http://quotes.toscrape.com"
response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser")
```

Aquest codi emet una sol·licitud HTTP GET a l'URL donat. Recupera les dades HTML que el servidor envia i emmagatzema aquestes dades en un objecte Python. Aquest és un codi HTML que mostra cometes, els seus autors i etiquetes. Cada cita està continguda dins d'un div amb la classe "quote" i conté el text de la cita, l'autor i un conjunt d'etiquetes. El text de la cita està contingut dins d'un interval amb la classe "text" i l'autor està contingut dins d'un petit element amb la classe "autor". Les etiquetes estan contingudes dins d'un div amb la classe "etiquetes" i cada etiqueta està continguda dins d'un element d'ancoratge amb la classe "etiqueta". Les etiquetes també contenen un metaelement amb la classe "keywords" que enumera les paraules clau associades a la cita.
El codi també inclou enllaços a les pàgines de l'autor, que tenen el format "/author/Author-Name".

Beautiful Soup és una biblioteca de Python per analitzar dades estructurades. Us permet interactuar amb HTML d'una manera similar a com interactueu amb una pàgina web mitjançant eines de desenvolupament. La biblioteca exposa un parell de funcions intuïtives que podeu utilitzar per explorar l'HTML que heu rebut.

Cerca elements per ID
En una pàgina web HTML, cada element pot tenir assignat un atribut id. Com el nom ja indica, aquest atribut id fa que l'element es pugui identificar de manera única a la pàgina. 

```python
quote = soup.find("span", class_="text").text
print(quote)
```

El codi ens ha imprès la primera cita del lloc web

```python
# Busca tots els elements HTML que contenen els noms dels autors
author_tags = soup.find_all("small", class_="author")

# Recorre la llista d'elements i extreu el text dels noms d'autors
authors = [author_tag.text for author_tag in author_tags]

# Imprimeix la llista d'autors
print(authors)
```

Aquest codi ens ha buscat tots els elements HTML que contenen la classe "author" i són de tipus "small", i extreura el text dels noms d'autors que es troben dins d'aquests elements. El resultat es una llista de noms d'autors.

```python
# Busca l'element HTML que conté les etiquetes "tag"
tags_container = soup.find("div", class_="tags")

# Busca tots els enllaços HTML dins de l'element "tags"
tag_links = tags_container.find_all("a", class_="tag")

# Recorre la llista d'enllaços i extreu el text de les etiquetes
tags = [tag_link.text for tag_link in tag_links]

# Imprimeix la llista d'etiquetes
print(tags)
```

Hem impres una llista amb els noms de les etiquetes "tag" que apareixen a la pàgina web.

```python
tag_name = 'thinking'
tag_url = f"{url}/tag/{tag_name}"
tag_response = requests.get(tag_url)
tag_soup = BeautifulSoup(tag_response.content, "html.parser")
quotes = tag_soup.find_all("span", class_="text")

for quote in quotes:
    print(quote.text)
```

Hem imprès totes les citacions que tenen la paraula "thinking".

```python
tags = soup.find_all("a", class_="tag")
for tag in tags:
    tag_url = url + tag["href"]
    tag_response = requests.get(tag_url)
    tag_soup = BeautifulSoup(tag_response.content, "html.parser")
    quotes = tag_soup.find_all("span", class_="text")
    for quote in quotes:
        author = tag_soup.find("small", class_="author").text
        if "Marilyn Monroe" in author:
            print(quote.text)
```

Hem buscat totes les etiquetes "a" amb la classe "tag",  el vincle per obtenir les pàgines que contenen les cites corresponents i, per cada pàgina, busca les frases amb l'etiqueta "span" i la classe "text". Després comprova si l'autor de la cita és "Marilyn Monroe" i, si és així, imprimeix la frase.

```python
from selenium import webdriver
from selenium.webdriver.common.by import By

url= "http://quotes.toscrape.com"
driver = webdriver.Chrome()  # o el driver que vulguis utilitzar
driver.get(url)
quote = driver.find_elements(By.XPATH, "//span[@class='text']")[0].text
print(quote)
```

```python
links = [link.get_attribute("href") for link in driver.find_elements(By.XPATH, "//a[@href]")]
print(links)
```

```python
love_quotes = driver.find_elements(By.XPATH, "//span[@class='text' and contains(text(), 'love')]")
for quote in love_quotes:
    print(quote.text)
```

```python
url1 = "https://es.wikipedia.org/wiki/Andy_Warhol"
page = requests.get(url1)
html= page.content
soup_warhol = BeautifulSoup(html, "html.parser")

# Troba el primer div amb l'identificador "mw-content-text"
content_div = soup_warhol.find("div", id="mw-content-text")

# Troba el primer paràgraf dins del div
first_paragraph = content_div.find("p")

# Imprimeix el text del paràgraf
print(first_paragraph.text)
```

```python
title = soup_warhol.title.string
print(title)
```

```python
velvet_section = soup_warhol.find('span', id='Música')

# Extreure la informació rellevant d'aquesta secció
info = velvet_section.find_next('p').get_text()

print(info)
```

SELENIUM 

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options as ChromeOptions
import time

op = webdriver.ChromeOptions()
op.add_argument('--enable-extensions')

wikiSelenium = webdriver.Chrome(options=op)
url1= "https://es.wikipedia.org/wiki/Andy_Warhol"
wikiSelenium.get(url1)
time.sleep(5)

# Obtenir informació del web, en aquest cas el títol de la  web
title = wikiSelenium.title
print(title)

# Busca l'enllaç a la secció "La Factory de Nueva York"
element = wikiSelenium.find_element(By.XPATH, "//a[@href='#La_Factory_de_Nueva_York']")

# Obté el contingut de l'enllaç
contingut_enllac = element.text

# Imprimeix el contingut de l'enllaç
print(contingut_enllac)
wikiSelenium.quit()
```

```python
# Inicialitzar el driver de selenium
driver = webdriver.Chrome('./chromedriver')

#    web
driver.get("https://es.wikipedia.org/wiki/Andy_Warhol")

time.sleep(10)

ul = driver.find_element("xpath", '//*[@id="mw-content-text"]')
data = []

# Buscar els enllaços dins de la llista y mostrar informació
enlaces = ul.find_elements(By.TAG_NAME, "a")
for enlace in enlaces:
    title = enlace.get_attribute("title")
    url = "https://es.wikipedia.org/wiki/Andy_Warhol" + enlace.get_attribute("href")
    text = enlace.text
    data.append([title, url])

    # Imprimir 
    print("Títol: ", title)
    print("URL: ", url)
    print("Text: ", text)
    print()
# Tancar el driver de selenium
driver.quit()
```
